```python
# 1. Look at your code from the lab data structures, and improve repeated code with loops.


products = ["t-shirt", "mug", "hat", "book", "keychain"] 
inventory = {
    "t-shirt": 10,
    "mug": 5,
    "hat": 8,
    "book": 20,
    "keychain": 5
}

customer_orders = []

print("Write 'done' when you're finished ordering.")

while True:
    order = input("Enter the name of a product to order (t-shirt, mug, hat, book, or keychain): ").strip().lower()
    
    if order == "done":
        break

    if order in products:
        if inventory[order] > 0:  
            customer_orders.append(order)
            inventory[order] -= 1 
            print(f"Added {order}. You have ordered {len(customer_orders)} product(s).")
        else:
            print(f"Sorry, {order} is out of stock.")
    else:
        print("That product is not available. Please choose from the given list.")


print("\nUpdated Inventory:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")


total_ordered = len(customer_orders)
total_products = len(products)
percentage_ordered = (total_ordered / total_products) * 100

order_status = (total_ordered, percentage_ordered)

print(f"Total Products Ordered: {order_status[0]}")
print(f"Percentage of Products Ordered: {order_status[1]:.2f}%")
```

```python
#3. Instead of updating the inventory by subtracting 1 from the quantity of each product, only do it for the products that were ordered (those in "customer_orders").

products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {
    "t-shirt": 10,
    "mug": 15,
    "hat": 8,
    "book": 20,
    "keychain": 5
}

customer_orders = []

print("Write 'done' when you're finished ordering.")

while True:

    order = input("Enter the name of a product to order (t-shirt, mug, hat, book, or keychain): ").strip().lower()

    if order == "done":
        break

    if order in products:
        if inventory[order] > 0: 
            customer_orders.append(order)  
            print(f"Added {order}. You have ordered {len(customer_orders)} product(s).")
        else:
            print(f"Sorry, {order} is out of stock.")
    else:
        print("That product is not available. Please choose from the given list.")

  
    another = input("Do you want to add another product? (yes/no): ").strip().lower()

    if another == "no":
        break


for product in customer_orders:
    if inventory[product] > 0:
        inventory[product] -= 1  

# Print the updated inventory
print("\nUpdated Inventory:")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")

# Print customer orders
print("\nYour Orders:")
for product in customer_orders:
    print(f"- {product}")
```
